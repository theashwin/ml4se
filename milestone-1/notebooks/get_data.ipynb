```python
# Get Java Code Dataset
import random
import pprint
import json

# Open File
with open('../data/java/java/final/jsonl/train/java_train_0.jsonl', 'r') as f:
    java_dataset = f.readlines()
size_of_dataset = len(java_dataset)

# Read data and store random 40 data points
target_dataset = []
no_of_datapoints = 100
attr_to_be_extracted = ['code', 'code_tokens', 'original_string']
for i in range(0, no_of_datapoints):
    rand_index = random.randint(0, size_of_dataset)
    java_code = json.loads(java_dataset[rand_index])
    target_code_data = {attr: java_code[attr] for attr in attr_to_be_extracted}
    target_dataset.append(target_code_data)
    try:
        with open('../data/javac/' + str(i+1) + '.java', 'w') as fp:
            fp.write(target_code_data['original_string'])
    except Exception as e:
        print(e)
        print("Error")
        pass
    
with open('../data/java.json', 'w') as fp:
    json.dump(target_dataset, fp)
pp = pprint.PrettyPrinter(indent=2)
pp.pprint(target_dataset)
```
